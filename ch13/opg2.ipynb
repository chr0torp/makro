```python
from statsmodels.tsa.filters.hp_filter import hpfilter
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = r'C:\Users\chris\Desktop\makro-Ã¸k\ch13\Chap 13_Exc 2_GDP data_US_UK_SV_DK.xlsx'
column_names = ['Year', 'US', 'UK', 'SV', 'DK']
df = pd.read_excel(file_path, sheet_name='Data', usecols='A,B,G,L,Q',names=column_names , skiprows=10, header=None, nrows=65)
# print(df.head())

```

```python
df['US'] = np.log(df['US'])
df['UK'] = np.log(df['UK'])
df['SV'] = np.log(df['SV'])
df['DK'] = np.log(df['DK'])
```

```python
cycle_us, trend_us = hpfilter(df['US'], lamb=100)
cycle_uk, trend_uk = hpfilter(df['UK'], lamb=100)
cycle_sv, trend_sv = hpfilter(df['SV'], lamb=100)
cycle_dk, trend_dk = hpfilter(df['DK'], lamb=100)

df['US_cycle'], df['US_trend'] = cycle_us, trend_us
df['UK_cycle'], df['UK_trend'] = cycle_uk, trend_uk
df['SV_cycle'], df['SV_trend'] = cycle_sv, trend_sv
df['DK_cycle'], df['DK_trend'] = cycle_dk, trend_dk

```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['US'], label='Original Data', color='blue')
plt.plot(df['Year'], df['US_trend'], label='Trend', color='red', linestyle='--')
plt.title('US GDP Decomposition')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['US_cycle'], label='Cycle', color='green', linestyle=':')
plt.title('US GDP Decomposition')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['DK'], label='Original Data', color='blue')
plt.plot(df['Year'], df['DK_trend'], label='Trend', color='red', linestyle='--')
plt.title('DK GDP Decomposition')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['DK_cycle'], label='Cycle', color='green', linestyle='-')
plt.title('DK GDP Decomposition')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend()
plt.grid(True)
plt.show()
```
